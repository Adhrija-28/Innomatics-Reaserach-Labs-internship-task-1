```python

# Problem 1: Unique Words in a Sentence

sentence = "python is easy and python is powerful"

words = sentence.split()
unique_words = set(words)

print("Problem 1")
print("Unique words count:", len(unique_words))
print("Unique words:", unique_words)
print()



# Problem 2: Highest Salary


employees = {
    "Ravi": 75000,
    "Anita": 68000,
    "Kiran": 72000
}

highest_name = ""
highest_salary = 0

for name, salary in employees.items():
    if salary > highest_salary:
        highest_salary = salary
        highest_name = name

print("Problem 2")
print("Highest Salary:", highest_name, "-", highest_salary)
print()



# Problem 3: Find Maximum and Minimum


numbers = [45, 22, 89, 10, 66]

maximum = numbers[0]
minimum = numbers[0]

for num in numbers:
    if num > maximum:
        maximum = num
    if num < minimum:
        minimum = num

print("Problem 3")
print("List:", numbers)
print("Max:", maximum)
print("Min:", minimum)
print()



# Problem 4: Count Products Above 1000


prices = [450, 1200, 899, 1500, 300]

count = 0

for price in prices:
    if price > 1000:
        count += 1

print("Problem 4")
print("Products above 1000:", count)
print()


# Problem 5: Calculate Attendance Percentage


attendance = ["P", "P", "A", "P", "P"]

present_count = 0

for day in attendance:
    if day == "P":
        present_count += 1

percentage = (present_count / len(attendance)) * 100

print("Problem 5")
print("Attendance Percentage:", percentage)
print()



# Problem 6: Remove Duplicate Phone Numbers


phone_numbers = [9876543210, 9123456789, 9876543210]

unique_numbers = set(phone_numbers)

print("Problem 6")
print("Unique phone numbers:", unique_numbers)
print()


# Problem 7: Count Character Frequency


text = "python"

frequency = {}

for char in text:
    if char in frequency:
        frequency[char] += 1
    else:
        frequency[char] = 1

print("Problem 7")
print("Character Frequency:", frequency)
print()



# Problem 8: Convert List to Tuple


list_values = [10, 20, 30]

tuple_values = tuple(list_values)

print("Problem 8")
print("Tuple:", tuple_values)
print()


# Problem 9: Check if Key Exists


employee_data = {
    "Ravi": 75000,
    "Anita": 68000
}

key_to_check = "Ravi"

print("Problem 9")
if key_to_check in employee_data:
    print("Employee exists")
else:
    print("Employee does not exist")
print()


# Problem 10: Calculate Average Marks


marks = [70, 75, 80, 65]

total = 0

for mark in marks:
    total += mark

average = total / len(marks)

print("Problem 10")
print("Average Marks:", average)
```
