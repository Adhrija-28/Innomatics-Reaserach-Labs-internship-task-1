```python
# -------------------- Problem 1 --------------------
class SmartLight:
    def __init__(self, name):
        self.name = name
        self.is_on = False

    def control(self, action):
        if action.upper() == "ON":
            self.is_on = True
        elif action.upper() == "OFF":
            self.is_on = False

        print(f"{self.name} is {'ON' if self.is_on else 'OFF'}")


# -------------------- Problem 2 --------------------
class Employee:
    def __init__(self, name, emp_id, department):
        self.name = name
        self.emp_id = emp_id
        self.department = department

    def display(self):
        print("Employee ID Card")
        print("Name:", self.name)
        print("ID:", self.emp_id)
        print("Department:", self.department)


# -------------------- Problem 3 --------------------
class Contact:
    def __init__(self, name, phone):
        self.name = name
        self.phone = phone

    def show(self):
        print("Contact Saved")
        print("Name:", self.name)
        print("Phone:", self.phone)


# -------------------- Problem 4 --------------------
class Product:
    def __init__(self, product_name, price):
        self.product_name = product_name
        self.price = price

    def print_tag(self):
        print("Product:", self.product_name)
        print("Price: ₹" + str(self.price))


# -------------------- Problem 5 --------------------
class Movie:
    def __init__(self, title, rating):
        self.title = title
        self.rating = rating

    def display(self):
        print("Movie:", self.title)
        print("Rating:", self.rating, "/ 5")


# -------------------- Problem 6 --------------------
class Delivery:
    def __init__(self, customer, address):
        self.customer = customer
        self.address = address

    def print_details(self):
        print("Delivery Details")
        print("Customer:", self.customer)
        print("Address:", self.address)


# -------------------- Main Execution --------------------

# Problem 1
light = SmartLight("Bedroom Light")
light.control("ON")
print()

# Problem 2
emp = Employee("Rahul", "EMP102", "IT")
emp.display()
print()

# Problem 3
contact = Contact("Anita", "9876543210")
contact.show()
print()

# Problem 4
product = Product("Headphones", 2499)
product.print_tag()
print()

# Problem 5
movie = Movie("Inception", 4.8)
movie.display()
print()

# Problem 6
delivery = Delivery("Suman", "Hyderabad")
delivery.print_details()
```
