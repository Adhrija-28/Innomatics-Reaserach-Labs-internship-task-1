```python
# Task 1: User Login Validation

print("Task 1: Login Validation ")

stored_credentials = {
    "username": "admin",
    "password": "1234"
}

user_try = {
    "username": "admin",
    "password": "1234"
}

login_status = (
    user_try["username"] == stored_credentials["username"]
    and
    user_try["password"] == stored_credentials["password"]
)

print("Login Successful" if login_status else "Invalid Credentials")



print("\n---------------------------------------\n")

# Task 2: Pass / Fail Result Evaluation


print(" Task 2: Result Evaluation ")

student_marks = [45, 78, 90, 33, 60]

pass_count = sum(1 for m in student_marks if m >= 50)
fail_count = len(student_marks) - pass_count

print(f"Total Students: {len(student_marks)}")
print(f"Students Passed: {pass_count}")
print(f"Students Failed: {fail_count}")


print("\n---------------------------------------\n")

# Task 3: Simple Data Cleaner
print("Task 3: Data Cleaning ")

raw_names = [" Ayesha ", "ram", " Charan "]

formatted_names = [person.strip().lower() for person in raw_names]

print("Cleaned Names:", formatted_names)




print("\n---------------------------------------\n")

# Task 4: Message Length Analyzer


print(" Task 4: Message Length Analyzer ")

chat_messages = ["Hi", "where are you from", "OK"]

for text in chat_messages:
    char_count = len(text)
    status = "Long Message" if char_count > 10 else "Short Message"
    print(f"'{text}' -> {char_count} characters ({status})")




print("\n---------------------------------------\n")

# Task 5: Error Message Detector

print("Task 5: Error Detector")

system_logs = ["INFO", "ERROR", "WARNING", "ERROR"]

error_total = system_logs.count("ERROR")

print(f"Number of ERROR logs: {error_total}")
```
